```python
# Read in the data

df = pd.read_csv(r'C:\Users\hecto\Documents\PROJECTS\SQL_Projects\1_Alex Freberg 𝘚𝘘𝘓 𝘋𝘢𝘵𝘢 𝘌𝘹𝘱𝘭𝘰𝘳𝘢𝘵𝘪𝘰𝘯 and 𝘋𝘢𝘵𝘢 𝘊𝘭𝘦𝘢𝘯𝘪𝘯𝘨 𝘪𝘯 𝘚𝘘𝘓\movies.csv')
df.head()
```

```python
# Check up for missing data
    
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col,pct_missing))
```

```python
df = df.dropna()
```

```python
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col,pct_missing))
```

```python
# Data types for our columns
df.dtypes
```

```python
# Delete decimals in Budget and Gross
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df.head()
```

```python
df = df.drop('yearcorrect', axis=1)
df.head()
```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Drop any duplicate values

df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
df.head()
```

```python
# Budget hight correlation
# Company will have a high correlation as well
# Scatter plot with budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Earnings')
plt.xlabel('Budget')
plt.ylabel('Gross')

plt.show()
```

```python
# Plt budget vs gross using seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color":"red"}, line_kws={"color":"blue"})
```

```python
# looking at Correlation

df.corr(method='pearson') #Pearson, Kendall, Spearman
```

```python
# high correlation between budget and gross
```

```python
correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
# Add quantitative value to qualitative value, in this case, Company column

df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized
```

```python
correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
# Votes and Budget have the highest correlation to gross earnings
# Company has low correlation

high_corr = sorted_pairs[(sorted_pairs) > 0.5]
high_corr
```
